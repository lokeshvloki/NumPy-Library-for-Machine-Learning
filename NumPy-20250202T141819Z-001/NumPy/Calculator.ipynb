```python
import numpy as np
```

```python
# Function to take matrix input
def input_matrix(name):
    rows = int(input(f"Enter the number of rows for {name}: "))
    cols = int(input(f"Enter the number of columns for {name}: "))
    print(f"Enter the elements of {name} row by row:")
    matrix = []
    for _ in range(rows):
        matrix.append(list(map(int, input().split())))
    return np.array(matrix)

# Menu-driven calculator
def matrix_operations():
    print("Matrix Operations Calculator using NumPy")
    print("1. Add Matrices")
    print("2. Subtract Matrices")
    print("3. Multiply Matrices")
    print("4. Find Determinant (Square Matrix Only)")
    print("5. Find Inverse of Matrix")
    print("6. Exit")

    choice = int(input("Choose an operation (1-5): "))
    if choice in [1, 2, 3]:
        matrix_a = input_matrix("Matrix A")
        matrix_b = input_matrix("Matrix B")
        if choice == 1:
            print("Result of Addition:\n", matrix_a + matrix_b)
        elif choice == 2:
            print("Result of Subtraction:\n", matrix_a - matrix_b)
        elif choice == 3:
            print("Result of Multiplication:\n", np.dot(matrix_a, matrix_b))
    elif choice == 4:
        matrix_a = input_matrix("Square Matrix A")
        if matrix_a.shape[0] == matrix_a.shape[1]:
            print("Determinant:\n", np.linalg.det(matrix_a))
        else:
            print("Determinant can only be calculated for square matrices!")
    elif choice == 5:
        matrix_a = input_matrix("Square Matrix A")
        if matrix_a.shape[0] == matrix_a.shape[1]:
            print("Inverse of Matrix is:\n", np.solve(matrix_a))
        else:
            print("Inverse can only be calculated for square matrices!")
    elif choice == 6:
        print("Exiting...")
    else:
        print("Invalid Choice!")

matrix_operations()
```
